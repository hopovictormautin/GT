```python
import pandas as pd
import requests
import lxml
from bs4 import BeautifulSoup as soup
from urllib.request import Request, urlopen #to override 403 error

data = {
    'Title': [],
    'Text':[],
    'Date':[],
    'Image':[] ,
    'Url':[],
    'Author':[]
}

def export_table_and_print(data):
    table = pd.DataFrame(data, columns = ['Title','Text','Date','Image','Url','Author'])

    table.to_csv('Green_Tech_Media_continue_.csv', index=False, encoding ='utf-8')   #0-2425, 2450-3250 = continue

def get_news_attributes(first_news):
    
    if first_news.find('h2',attrs = {'class': 'general-heading'}):
        title = first_news.find('h2',attrs = {'class': 'general-heading'}).text   
        data['Title'].append(title)
    else:
        data['Title'].append('No Result')
        
    if first_news.find('p').contents[0]:
        text = first_news.find('p').contents[0]
        data['Text'].append(text)
    else:
        data['Text'].append('No Result')
        
    if first_news.find('time'):
        date = first_news.find('time')['datetime']
        data['Date'].append(date)
    else:
        data['Date'].append('No Result')
    
    if first_news.find('span',attrs = {'class':"by-info"}):
        author = first_news.find('span',attrs = {'class':"by-info"}).text
        data['Author'].append(author)
    else:
        data['Author'].append('No Result')

    if  first_news.find('img', attrs = {'class':"lazy"}):
        image = first_news.find('img', attrs = {'class':"lazy"})['data-original']
        data['Image'].append(image)
    else:
        data['Image'].append('No Result')
        
        
    if first_news.find('div',attrs = {'class':"article-wrapper"}):
        url = 'https://www.greentechmedia.com' + first_news.find('div',attrs = {'class':"article-wrapper"}).find('a')['href']
        data['Url'].append(url)
    else:
        data['Url'].append('No Result')
        
#     title = first_news.find('h2',attrs = {'class': 'general-heading'}).text   
#     text = first_news.find('p').contents[0]
#     date = first_news.find('time')['datetime']
#     author = first_news.find('span',attrs = {'class':"by-info"}).text
#     url = 'https://www.greentechmedia.com' + first_news.find('div',attrs = {'class':"article-wrapper"}).find('a')['href']
#     image = first_news.find('img', attrs = {'class':"lazy"})['data-original']
    
    
#     data['Title'].append(title)
#     data['Text'].append(text)
#     data['Date'].append(date)
#     data['Image'].append(image)
#     data['Url'].append(url)
#     data['Author'].append(author)

page_numbers = range(2450,3250,25)    #0-2425, 2450-3250 = continue
search_url = "https://www.greentechmedia.com/articles/category/storage/"
urls_to_search = []
#urls_to_search.append(search_url)

for page_number in page_numbers:
    new_search_url= search_url + 'page' + '/' + 'P' +str(page_number)+'/'
    urls_to_search.append(new_search_url)
print(len(urls_to_search))
print(urls_to_search)

k = 0
for p_s in urls_to_search:
    print(k)
    search_url = p_s 

    req = Request(search_url, headers = {'User-Agent':'Mozilla/5.0'})
    page =urlopen(req).read()
    
    page_soup = soup(page,'html.parser')

    news_page = page_soup.find_all('div', attrs = {'class': 'article-list-holder home'})

    #print(news_page)
    for first_news in news_page:

        get_news_attributes(first_news)
        print (data)
 
        export_table_and_print(data)
    k += 1
```
